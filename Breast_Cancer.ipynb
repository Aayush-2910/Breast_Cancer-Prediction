The Breast Cancer dataset contains 30 numerical input features that are extracted from digitized images of cell nuclei taken from breast tissue sampleslicized text

**Main Physical Characteristics**

| Feature           | Description                                          |
| ----------------- | ---------------------------------------------------- |
| Radius            | Distance from the center to the boundary of the cell |
| Texture           | Variation in gray-scale intensity                    |
| Perimeter         | Length of the cell boundary                          |
| Area              | Area of the cell                                     |
| Smoothness        | Smoothness of the cell boundary                      |
| Compactness       | How tightly packed the cell is                       |
| Concavity         | Depth of inward curves in the boundary               |
| Concave Points    | Number of concave portions                           |
| Symmetry          | Symmetry of the cell shape                           |
| Fractal Dimension | Complexity of the cell boundary                      |


**Target Variable**
| Value | Meaning                |
| ----- | ---------------------- |
| 1     | Malignant (Cancerous)  |
| 0     | Benign (Non-cancerous) |

**Why These Features Are Important**

Cancerous cells usually have:
* Larger size
* Irregular shape
* More concave boundaries
* Less symmetry

These features help machine learning models distinguish between malignant and benign tumors
